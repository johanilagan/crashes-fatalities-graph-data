```python
import pandas as pd
```

```python
df = pd.read_csv('Project2_Dataset_Corrected.csv')
df.head(5)
```

```python
df.dtypes
```

```python
df['Bus Involvement'] = df['Bus Involvement'].replace({'Yes': True, 'No': False})
df['Heavy Rigid Truck Involvement'] = df['Heavy Rigid Truck Involvement'].replace({'Yes': True, 'No': False})
df['Articulated Truck Involvement'] = df['Articulated Truck Involvement'].replace({'Yes': True, 'No': False})
df['Heavy Rigid Truck Involvement'] = df['Heavy Rigid Truck Involvement'].replace({'Yes': True, 'No': False})
df['Articulated Truck Involvement'] = df['Articulated Truck Involvement'].replace({'Yes': True, 'No': False})

df['Christmas Period'] = df['Christmas Period'].replace({'Yes': True, 'No': False})
df['Christmas Period'] = df['Christmas Period'].astype(bool)
df['Easter Period'] = df['Easter Period'].replace({'Yes': True, 'No': False})
df['Easter Period'] = df['Easter Period'].astype(bool)
```

```python
df = df.rename(columns={'Crash ID': 'crash_id', 
                                  'State': 'state', 
                                  'Month':'month', 
                                  'Year':'year',
                                  'Dayweek':'day_of_week',
                                  'Time':'time',
                                  'Crash Type':'crash_type',
                                  'Number Fatalities':'number_fatalities',  
                                  'Bus Involvement':'bus_involved',
                                  'Heavy Rigid Truck Involvement':'heavy_rigid_truck_involved',
                                  'Articulated Truck Involvement':'articulated_truck_involved',
                                  'Speed Limit':'speed_limit',
                                  'National Remoteness Areas':'remoteness_area',
                                  'SA4 Name 2021':'sa4_name',
                                  'National LGA Name 2024':'lga_name',
                                  'National Road Type':'road_type',
                                  'Day of week':'weekday',
                                  'Time of day':'time_of_day',
                                  'Road User':'road_user',
                                  'Gender':'gender',
                                  'Age':'age',
                                  'Age Group':'age_grp',
                                  'Christmas Period':'christmas_period',
                                  'Easter Period':'easter_period'
                                 })

df.head(5)
```

```python
duplicates = df[df.duplicated()]
duplicates
```

```python
crash_df = df[['crash_id', 'crash_type', 'number_fatalities', 'speed_limit']].drop_duplicates()
crash_df
```

```python
# Create datetime node/df
datetime_df = df[["day_of_week","month","year", "time", "weekday", "time_of_day", "christmas_period", "easter_period"]].drop_duplicates()
d_index = range(1,1+len(datetime_df))
datetime_df["date_id"] = ["D" + str(i) for i in d_index]
datetime_df = datetime_df[["date_id", "day_of_week","month","year", "time", "weekday", "time_of_day", "christmas_period", "easter_period"]]
datetime_df
```

```python
# Create person node/df
person_df = df[["road_user","gender","age", "age_grp"]].drop_duplicates()
p_index = range(101,101+len(person_df))
person_df["person_id"] = ["P" + str(i) for i in p_index]
person_df = person_df[["person_id","road_user","gender","age", "age_grp"]]
person_df
```

```python
# Create location node/df
location_df = df[["state","remoteness_area","sa4_name", "lga_name", "road_type"]].drop_duplicates()
loc_index = range(1,1+len(location_df))
location_df["location_id"] = ["L" + str(i) for i in loc_index]
location_df = location_df[["location_id","state","remoteness_area","sa4_name", "lga_name", "road_type"]]
location_df
```

```python
# Create vehicle type node/df
vehicle_type_df = df[["bus_involved","heavy_rigid_truck_involved","articulated_truck_involved"]].drop_duplicates()
v_index = range(1,1+len(vehicle_type_df))
vehicle_type_df["vehicle_id"] = ["V" + str(i) for i in v_index]
vehicle_type_df = vehicle_type_df[["vehicle_id","bus_involved","heavy_rigid_truck_involved","articulated_truck_involved"]]
vehicle_type_df
```

```python
df = df.merge(datetime_df, on=["day_of_week","month","year", "time", "weekday", "time_of_day", "christmas_period", "easter_period"], how='left')
df = df.merge(location_df, on=["state","remoteness_area","sa4_name", "lga_name", "road_type"], how='left')
df = df.merge(person_df, on=["road_user","gender","age", "age_grp"], how='left')
df = df.merge(vehicle_type_df, on=["bus_involved","heavy_rigid_truck_involved","articulated_truck_involved"], how='left')
df
```

```python
rel_involved = df[['crash_id','person_id']]
rel_involved
```

```python
rel_occurredAt = df[['crash_id', 'location_id']]
rel_occurredAt
```

```python
rel_involvedVehicle = df[['crash_id','vehicle_id']]
rel_involvedVehicle
```

```python
rel_occurredOn = df[['crash_id','date_id']]
rel_occurredOn
```

```python
csv_name = ['crash.csv','datetime.csv','person.csv','location.csv','vehicle_type.csv','rel_involved.csv','rel_occurredAt.csv','rel_involvedVehicle.csv','rel_occurredOn.csv']
dfs = [crash_df, datetime_df, person_df, location_df, vehicle_type_df, rel_involved, rel_occurredAt, rel_involvedVehicle, rel_occurredOn]
for i in range(len(csv_name)):
    dfs[i].to_csv(csv_name[i], index = False)
```
